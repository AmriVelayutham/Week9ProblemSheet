# Recursion

A recursive function is a function that calls itself (a process called recursion). Functions can be directly
recursive (if they directly call themselves) or indirectly recursive (if function `a()` calls function `b()`, which calls
function `a()`). Recursion is a very common way to solve a wide range of computational problems, but it can be hard to
understand how a recursive function comes to the solution it comes to. It is also easy to write a recursive
function that never produces a result (because it gets stuck in a never ending loop).

## Problem 44:

We can define the factorial function recursively thus:

$$
\mathrm{fact}(n) = 
\begin{cases} 
      1 & n = 0 \\
      n . \mathrm{fact}(n-1)  & \mathrm{otherwise} 
   \end{cases}
$$

A recursive implementation of the factorial function is presented below. It includes print statements to 
how it is called when it runs. In the following cell explore how the function works with different arguments
and answer the questions.

````python
def fact(n):
    print("fact called with argument ", n)
    if n == 0:
        print("Base case - returning", 1)
        return 1
    else:
        sol = n * fact(n-1)
        print("Recursive case - returning", sol)
        return sol

````

````python
fact(4)
````

## Problem 45:

The Euclidean algorithm provides a recursive method to calculate the greatest common divisor of two integers. The
greatest common divisor function, $\mathrm{gcd}(p,q)$ is defined:

$$
\mathrm{gcd}(p,q) = 
\begin{cases} 
      p & \mathrm{if}\, q = 0 \\
      \mathrm{gcd}(q, p\mathrm{\%}q)  & \mathrm{if}\, q > 0 
   \end{cases}
$$

where $p\mathrm{\%}q$ is the remainder of $\frac{p}{q}$ (the `%` operator in Python). Implement the Euclidean algorithm in the cell below. Make sure you keep the $p$'s and $q$'s the right way around. What is the greatest common divisor of 10 and 5?

````python
# Write a solution here
def gcd(p, q):
    if q == 0:
        return p
    elif q>0:
        return gcd(q, p % q)  #why is it gcd (q, p%q), why modulo here
# This should return 5.
gcd(10, 5)
````

## An introduction to NumPy 

We looked at NumPy in arrays in class. Briefly NumPy is a python module that adds a new way of storing data (a NumPy array) and some
functions that work on these arrays. Functions can either work element wise on arrays (ufuncs) or can operate on the whole array 
(for example to return the maximum value, or to do matrix multiplication of two arrays such that the value of each output element
depends on many input elements. The key differences between arrays and lists are:

* all elements in an array must be of the same type (e.g. all integers or all floating point numbers) but lists can have different element types.
* arrays can be multi-dimensional (lists are always one dimensional), so arrays can have multiple indexes.
* working with whole arrays (e.g. using ufuncs or operators like `+` or `*`) is fast because it avoids looping.

If you have used Matlab before, you will find arrays very similar to the default data structure in that language.

To end this weeks problems we'll look to see how we can write functions of whole arrays, and then look at
how we can combine NumPy arrays, functions and plotting together to make some graphs of simple mathematical
functions.


## Problem 46:

Write a function, `rescale()` that takes an array as input and returns a corresponding array of values 
scaled to lie in the range 0.0 to 1.0.

*Hint:* If `L` and `H` are the lowest and highest values in the original array, then the replacement for a value `v` should be `(v-L) / (H-L)`.)


````python
# Create your solution here
import numpy as np

def rescale (array):
    """
    Rescale an array such that the values lie between 0.0 and 1.0
    """
    L=np.min(array)
    H=np.max(array)
    rescaled_array= (array-L)/(H-L)
    return rescaled_array

# Example of use - see problem 47
v = np.arange(0,100,2)
v_resc = rescale(v) 
print(v_resc)
print(v)
print(len(v))
````

## Problem 48.4:

The derivative of a polynomial is given by:

$$\frac{d}{dx}f(x) = f^{\prime}(x) = \sum_{k=0}^m k c_k x^{k-1}$$

In the next two cells create a function, `polyprime`, that returns the derivative of a polynomial, then create two plots which show a polynomial and its derivative, one above the other.

````python
# implement the polyprime function here
def poly(x,coefs):
    y=0
    for i, coefs in enumerate(coefs):
        y += coefs* x**i
    return y

def polyprime(x,coefs):
    yprime=0
    for i, coefs in enumerate(coefs):
        yprime += i*coefs* x**(i-1)
    return yprime
````

````python
# make a graphs here
fig, ax = plt.subplots(nrows=2, ncols=1)
ax[0].plot(np.linspace(-1, 1), poly(np.linspace(-1, 1), [2, 4, 3]), label = 'Parabola')
ax[1].plot(np.linspace(-1, 1), polyprime(np.linspace(-1, 1), [2, 4, 3]), label = 'Derivative')
fig.tight_layout()
plt.show()
````
